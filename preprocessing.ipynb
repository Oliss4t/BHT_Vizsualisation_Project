```python
import pandas as pd
import pycountry
import missingno as msno
```

## Preprocessing Each Data into Same Format Columns

```python
data2015 = pd.read_csv('./data/happy/2015.csv')
data2015['Year'] = 2015

data2016 = pd.read_csv('./data/happy/2016.csv')
data2016['Year'] = 2016

data2017 = pd.read_csv('./data/happy/2017.csv')
data2017 = data2017.rename(columns = {'Happiness.Rank':'Happiness Rank', 'Happiness.Score' : 'Happiness Score',
                                      'Economy..GDP.per.Capita.' : 'Economy (GDP per Capita)', 'Health..Life.Expectancy.' : 'Health (Life Expectancy)',
                                      'Trust..Government.Corruption.' : 'Trust (Government Corruption)', 'Dystopia.Residual' : 'Dystopia Residual'})
data2017['Year'] = 2017


data2018 = pd.read_csv('./data/happy/2018.csv')
data2018 = data2018.rename(columns = {'Overall rank':'Happiness Rank', 'Country or region' : 'Country', 'Score' : 'Happiness Score',
                                      'GDP per capita' : 'Economy (GDP per Capita)', 'Social support' : 'Family',
                                      'Healthy life expectancy' : 'Health (Life Expectancy)','Freedom to make life choices' : 'Freedom',
                                      'Perceptions of corruption' : 'Trust (Government Corruption)'})
data2018['Year'] = 2018


data2019 = pd.read_csv('./data/happy/2019.csv')
data2019 = data2019.rename(columns = {'Overall rank':'Happiness Rank', 'Country or region' : 'Country', 'Score' : 'Happiness Score',
                                      'GDP per capita' : 'Economy (GDP per Capita)', 'Social support' : 'Family',
                                      'Healthy life expectancy' : 'Health (Life Expectancy)','Freedom to make life choices' : 'Freedom'
                                     , 'Perceptions of corruption' : 'Trust (Government Corruption)'})
data2019['Year'] = 2019

data2020 = pd.read_csv('./data/happy/2020.csv')
data2020['Happiness Rank'] =  range(1, len(data2020.index)+1)
data2020 = data2020.rename(columns = {'Country name' : 'Country', 'Ladder score' : 'Happiness Score',
                                      'Logged GDP per capita' : 'Economy (GDP per Capita)', 'Social support' : 'Family', 'Healthy life expectancy' : 'Health (Life Expectancy)',
                                      'Freedom to make life choices' : 'Freedom', 'Perceptions of corruption' : 'Trust (Government Corruption)'})
data2020['Year'] = 2020

data2021 = pd.read_csv('./data/happy/2021.csv')
data2021['Happiness Rank'] =  range(1, len(data2021.index)+1)
data2021 = data2021.rename(columns = {'Country name' : 'Country', 'Ladder score' : 'Happiness Score',
                                      'Logged GDP per capita' : 'Economy (GDP per Capita)', 'Social support' : 'Family', 'Healthy life expectancy' : 'Health (Life Expectancy)',
                                      'Freedom to make life choices' : 'Freedom', 'Perceptions of corruption' : 'Trust (Government Corruption)'})
data2021['Year'] = 2021

data2022 = pd.read_csv('./data/happy/2022.csv')
data2022['Happiness Rank'] =  range(1, len(data2022.index)+1)
data2022 = data2020.rename(columns = {'Country name' : 'Country', 'Ladder score' : 'Happiness Score',
                                      'Logged GDP per capita' : 'Economy (GDP per Capita)', 'Social support' : 'Family', 'Healthy life expectancy' : 'Health (Life Expectancy)',
                                      'Freedom to make life choices' : 'Freedom', 'Perceptions of corruption' : 'Trust (Government Corruption)'})
data2022['Year'] = 2022

data2022
```

## Store Each Data from to 2015 - 2022 into New Data Frame

```python
datarank = pd.DataFrame(columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                                   'Family', 'Health (Life Expectancy)', 'Freedom','Trust (Government Corruption)',
                                   'Generosity', 'Year'])
n = [data2015, data2016, data2017, data2018, data2019, data2020, data2021,  data2022]
for i in n:
    datarank = datarank.append(i[['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                                   'Family', 'Health (Life Expectancy)', 'Freedom','Trust (Government Corruption)',
                                   'Generosity', 'Year']], ignore_index = 'True')
```

```python
df_country_region_mapping_2022 = data2022.loc[:, ['Country', 'Regional indicator']].rename(columns = {'Country' : 'Country', 'Regional indicator' : 'Region 1'})
df_country_region_mapping_2015 = data2015.loc[:, ['Country', 'Region']].rename(columns = {'Country' : 'Country', 'Region' : 'Region 2'})
df_country_region_mapping_2016 = data2016.loc[:, ['Country', 'Region']].rename(columns = {'Country' : 'Country', 'Region' : 'Region 3'})

df_country_region_mapping = pd.merge(df_country_region_mapping_2022, df_country_region_mapping_2015, how='left', on=["Country"])
df_country_region_mapping = pd.merge(df_country_region_mapping, df_country_region_mapping_2016, how='left', on=["Country"])
df_country_region_mapping['Region'] =df_country_region_mapping[['Region 1', 'Region 2', 'Region 3']].bfill(axis=1).iloc[:, 0]
df_country_region_mapping.drop(['Region 1', 'Region 2', 'Region 3'], axis=1, inplace=True)
df_country_region_mapping
```

```python
countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_2

df_country_region_mapping['Country Code'] = df_country_region_mapping.apply(lambda row : countries.get(row['Country'], 'Unknown code'), axis = 1)
df_country_region_mapping
```

```python
pd_happy = pd.merge(datarank, df_country_region_mapping, how='left', on=["Country"])
```
